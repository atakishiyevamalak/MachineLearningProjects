```python
import pandas as pd
```

```python
df.drop(['person_id', 'visits_last_year'], axis=1, inplace=True)
```

```python
len(df.columns)
```

```python
df['annual_medical_cost'].isna().sum()
```

```python
X = df.drop(['annual_medical_cost'], axis=1)
y = df['annual_medical_cost'].copy()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median'))
])
```

```python
num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns
```

```python
transformer = ColumnTransformer([
    ('cat', cat_pipeline, cat_features),
    ('num', num_pipeline, num_features)
], remainder='passthrough')
```

```python
X_train_transformed = transformer.fit_transform(X_train)
X_train_transformed
```

```python
columns = transformer.get_feature_names_out()
```

```python
X_train_transformed_df = pd.DataFrame(data=X_train_transformed, columns = columns)
```

```python
X_test_transformed = transformer.transform(X_test)
X_test_transformed
```

```python
X_test_transformed_df = pd.DataFrame(data=X_test_transformed, columns = columns)
```

```python
X_test_transformed_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
rfc = RandomForestRegressor()
```

```python
lr.fit(X_train_transformed_df, y_train)
```

```python
print(f'train score: {lr.score(X_train_transformed_df, y_train)}\ntest score: {lr.score(X_test_transformed_df, y_test)}')
```

```python
rfc.fit(X_train_transformed_df, y_train)
```

```python
print(f'train score: {rfc.score(X_train_transformed_df, y_train)}\ntest score: {rfc.score(X_test_transformed_df, y_test)}')
```
